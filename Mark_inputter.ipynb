### Question 01
Suppose, you are a developer of a software company named 'Hacker Man'.Your supervisor asked you to develop a password generator program, for MIPS architecture, that can both generate passwords and break them. After building   the   program,   you   observed   that,   the   program   was   taking  $X$ milliseconds to execute and the password generation was taking 90% ofthe total execution time (Here $X$ is your **BRACU ID**. For example, if your ID is **12456789** then the time is **12456789 milliseconds**). After reporting  to your boss fearfully, he said it was unacceptable andthreatened   to   fire   you   unless   you   can   increase   the   whole   program'sperformance by a factor of 5 by optimizing the password generation. Now, your job's future rest in your optimization skills and the results of your analysis. Let's start with the analysis.

**Ans:**
This question can be solved without using the BRACU ID.
\begin{aligned}
    \textit{Improvement} &= \frac{1}{(1-p)+\frac{p}{n}}\\
    5 &= \frac{1}{(1-0.9)+\frac{0.9}{n}}\\
    n &= 9
\end{aligned}

In the meantime, you are thinking if you can decrease the execution time of the password breaking operation. You thought of a slight optimization of that operation and made some rough calculations. You optimized your algorithm to have an average CPI of Y and instruction count of $Y\times10^7$ (Here $Y$ is the third digit of your **BRACU ID**. ### Question 3
Suppose you are running the SPEC CINT2006 benchmarking program, and the program ran three separate programs listed below:

| Program name| Instruction count ($\times 10^9$)| Clock rate (GHz) | CPI | Execution time | Reference time | SPECratio |
| --- | --- | --- | --- | --- | --- | --- |
| BFG | P+1 | 2.4 | 2.5 | ? | 86 | ? |
| Vega | Q+1 | 2.4 | 2.0 | ? | 12 | ? |
| Checker | R+1 | 2.4 | 3.0 | ? | 15 | ? |

Here **P, Q, R** are the **6th, 7th, and 8th** digit of your **BRACU ID**.

```python
import gspread
import pandas as pd
from oauth2client.service_account import ServiceAccountCredentials
from numpy import prod
from colorama import Fore
from colorama import Style

# define the scope
scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']

# add credentials to the account - took out my private key since I don't want API calls
creds = ServiceAccountCredentials.from_json_keyfile_name('', scope)

# authorize the clientsheet 
client = gspread.authorize(creds)

sheet = client.open('')

# get the first sheet of the Spreadsheet
sheet_instance = sheet.get_worksheet(2)

#Case 1: Execution time = T_unaffected (time for password breaking operation or 10% of the BRACU ID time)
def solution(id):

    cpi = int(id[2])
    ic = int(id[2]) * 10**7

    cpu_time = (int(id)*0.1) * 10**(-3) #converting to second

    rate_t_unaffected = (ic * cpi) / cpu_time

    #Case 2: Execution time = Original execution time (BRACU ID milliseconds)
    cpu_time = (int(id)) * 10**(-3)

    rate_full_id = (ic * cpi) / cpu_time
    #Case 3: Execution time = Optimized execution time (BRACU ID / 5 milliseconds)
    cpu_time = (int(id)/5) * 10**(-3)

    rate_optimizedby5 = (ic * cpi) / cpu_time


    p = int(id[5])
    q = int(id[6])
    r = int(id[7])

    ratio = {}

    ratio['BFG'] = 86 / (((p + 1) * 2.5) / 2.4)
    ratio['Vega'] = 12 / (((q + 1) * 2.0) / 2.4)
    ratio['Checker'] = 15 / (((r + 1) * 3.0) / 2.4)

    print("BRACU ID: ", id)
    print()
    print("Answers for Question 1: 9")
    print()
    print("Answers for Question 2: ")
    print()
    print(f'CPU Clock Rate (For case 1) giving t_unaffected:\t \t{rate_t_unaffected/1000} KHz')
    print()
    print(f'CPU Clock Rate (For case 2) using full execution time: \t \t{rate_full_id/1000} KHz')
    print()
    print(f'CPU Clock Rate (For case 3) by optmizing by 5: \t \t \t{rate_optimizedby5/1000} KHz')
    print()
    print("Answers to Question 3: ")


    for key in ratio.keys():
      print(f'SPECratio of {key} is {ratio[key]}')

    geo_mean = prod(list(ratio.values())) ** (1. / 3)

    print(f'Geometric Mean: {geo_mean}')

    
def input_marks(id):
    marks = input(f"What marks did you give this person with id-{id} ")
    cell = sheet_instance.find(id)
    sheet_instance.update_cell(cell.row,11, marks)
    print(f"The marks for {id}, is {marks}")
```

```python
while id!=-100:
    print("-----------------------------------------")
    id = input('Enter BRACU ID: ')
    if int(id) == -100:
        break
    solution(id)
    
    try:
        input_marks(id)
    except:
        print(f"{Fore.RED}Whoopsie Daisy{Style.RESET_ALL}")
```
